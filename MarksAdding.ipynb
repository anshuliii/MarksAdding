**Marks Adding :** Input File contains Roll Nos. and Marks. Output File adds all the marks.

```python
import pandas as pd

df = pd.read_csv('/content/Mini Project - Marks Adding.csv')

print("Input Data:")
print(df)
print(f"\nTotal records in input: {len(df)}")

output_df = df.groupby('Roll Num')['Marks'].sum().reset_index()

print("\nOutput Data (After adding marks for duplicate roll numbers):")
print(output_df)
print(f"\nTotal unique roll numbers: {len(output_df)}")

output_df.to_csv('output_marks.csv', index=False)

print("\nOutput file 'output_marks.csv' has been created successfully!")

duplicate_rolls = df[df.duplicated(['Roll Num'], keep=False)]['Roll Num'].unique()
if len(duplicate_rolls) > 0:
    print(f"\nRoll numbers with multiple entries: {duplicate_rolls}")
    print("\nSummary of mark additions:")
    for roll in duplicate_rolls:
        original_marks = df[df['Roll Num'] == roll]['Marks'].tolist()
        total_marks = sum(original_marks)
        print(f"Roll {roll}: {' + '.join(map(str, original_marks))} = {total_marks}")
else:
    print("\nNo duplicate roll numbers found in the input data.")
```
